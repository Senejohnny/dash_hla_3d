```python
import pandas as pd
import numpy as np
```

```python
desa = pd.read_pickle('../data/desa_3d_view.pickle')
```

```python
ep_db = pd.read_pickle('../data/20201105_EpitopevsHLA_distance.pickle')
```

```python
elliporo_filter = ['High',]
ind_filt = ep_db['ElliPro Score'].apply(lambda x: x in elliporo_filter)
ep_ellipro = ep_db[ind_filt].Epitope.values.tolist()
```

```python
desa['EpvsHLA_Donor_filtered'] = desa.EpvsHLA_Donor.apply(lambda x: {key:value for key, value in x.items() if key in ep_ellipro})
```
